```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python

reviews_df = pd.read_csv('../data/2022-09-15/reviews.csv')

#reviews_df['date'] = pd.to_datetime(reviews_df['date'], format=r'%Y-%m-%d')

print(reviews_df.shape)

display(reviews_df.head())

reviews_df_cnt = reviews_df.groupby(by='date').size()
```

```python
reviews_df_cnt
```

```python
reviews_df_cnt_subset = reviews_df_cnt.loc['2018-01-01':'2022-08-01']
#dates for about 4.5 years. trimmed some of the end data to account for reviews lagging as a result of people not immediately leaving reviews
```

```python
reviews_df_cnt_subset.info()
```

```python
f = plt.figure()
f.set_figwidth(40)
f.set_figheight(20)

plt.plot(reviews_df_cnt_subset.index, reviews_df_cnt_subset)
plt.show()
#clearly shows a trend of more booking since covid
#weekend peaks are much higher
#weekday books are also higher but not as much (my assumption is that people still have to work on weekdays so there's not as much unrestrained room for more "vacationing")
```
